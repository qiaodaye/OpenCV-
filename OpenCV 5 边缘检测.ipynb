### Canny 边缘检测
- 1）使用高斯滤波器，以平滑图像滤除噪声
- 2）计算出每个像素点的梯度强度和方向
- 3）应用非极大值（Non-Maximun Suppression）抑制，以消除边缘检测带来的杂散影响
- 4）应用双阈值（Double-Threshold)检测来确定真实的和潜在的边缘
- 5）通过抑制孤立的弱边缘最终完成边缘检测

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline #相当于展示的更加方便 不用再show出来了
```

```python
img = cv2.imread('/Users/qiaoye/Desktop/7.4.jpeg')

```

```python
v1 = cv2.Canny(img,80,125)
plt.imshow(v1)
```
