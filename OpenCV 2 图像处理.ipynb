### 图形阈值

**res,dst = cv2.threshold(src,thresh,maxval,type)**
- src: 输入层，只能输入单通道图像，通常来说为灰度图
- dst: 输出图
- thresh：阈值
- maxval：当像素值超过了阈值（或者小于阈值，根据type来决定），所赋予的值
- type： 二值化操作类型，包括以下5种类型：
- cv2.THRESH_BINARY 超过阈值部分取maxval（最大值），否则取0
- cv2.THRESH_BINARY_INV  是THRESH_BINARY的反转
- cv2.THRESH_TRUNC 大于阈值部分设置为阈值，否则不变
- cv2.THRESH_TOZERO 大于阈值部分设置为不变，否则设置为0
- cv2.THRESH_TOZEROS_INV  THRESH_TOZEROS 的反转

### 图像平滑

 - 均值滤波  简单的平均卷积操作
 - 方框滤波  基本和均值滤波一样，可以选择归一化
 - 高斯滤波  高斯卷积中的数值满足高斯分布，相当于更重视中间的
 - 中值滤波  相当于用中值代替

### 均值滤波

```python
plt.imshow(img)
```

```python
blur = cv2.blur(img,(3,3)) 
plt.imshow(blur)
```

### 方框滤波

```python
box = cv2.boxFilter(img,-1,(3,3),normalize = True) #true 的情况和均值滤波形式是一样的
plt.imshow(box)
```

```python
box = cv2.boxFilter(img,-1,(2,2),normalize = False) #true 的情况和均值滤波形式是一样的
plt.imshow(box)
```

### 高斯滤波

```python
aussim = cv2.GaussianBlur(img,(15,15),1)
plt.imshow(aussim)
```

### 中值滤波

```python
median = cv2.medianBlur(img,5)
plt.imshow(median)
```
